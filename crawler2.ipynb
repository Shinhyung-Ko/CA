```python
from bs4 import BeautifulSoup as bs
```

```python
fd = rq.get('https://finance.naver.com/item/sise.naver?code=005930')
soup = bs(fd.content, 'html.parser')
```

```python
print(soup.find_all('strong',id="_nowVal")[0].text)
```

```python
def 주식(x):
    fd = rq.get(f'https://finance.naver.com/item/sise.naver?code={x}')
    soup = bs(fd.content, 'html.parser')
    return soup.find_all('strong',id="_nowVal")[0].text

```

```python
종목들 = ['005930', '066575', '005380', '035720', '034220', '003490']

현재가 = open('현재가.txt', 'w')
for i in 종목들:
    현재가.write(주식(i))
    현재가.write("원\n")


현재가.close()
```

```python
for i in 종목들:
  print(i)
```
