```python
import tensorflow as tf

```

```python
#키 = [170, 180, 175, 165]
#신발 = [260, 270, 265, 255]
키 = 170
신발 = 260
```

```python
#신발 = 키 * a + b
a = tf.Variable(0.1)
b = tf.Variable(0.2)


def 손실함수():
    #return ts.square<-안에 값 제곱(실제값 - 예측값)
    예측값 = 키 * a + b
    return tf.square(260 - 예측값)

opt = tf.keras.optimizers.legacy.Adam(learning_rate=0.1)

for i in range(300):
    opt.minimize(손실함수, var_list=[a,b])
    print(a,b)
```
