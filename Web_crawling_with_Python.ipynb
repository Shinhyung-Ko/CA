```python
#file_read/write
파일 = open('a.txt', 'w')#'w' = 없으면 만들어라 <- 갈아치우기 #wb - 바이너리형식(0,1)ex)이미지파일
파일.write('hello') #덮어쓰는 문법. 새로 생성!
파일.close()
#-----------------------------

파일 = open('a.txt', 'a') #'a' 추가모드
파일.write('\n반가웡')
파일.close

파일 = open('a.txt', 'r')
print(파일.read())
파일.close()
```
