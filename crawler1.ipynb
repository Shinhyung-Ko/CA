```python
import requests as rq #파이썬으로 웹 접속
from bs4 import BeautifulSoup as bs#파이썬으로 html문서 분석
```

```python
#rq.get('URL') 웹 접속 방법
data = rq.get('https://finance.naver.com/')
```

```python
print(data)
#response [200] -> 잘 가져왔다
print(data.content)
print(data.status_code) #200이 뜨면 잘 가져온거임
```

```python
soup = bs(data.content, 'html.parser')
#print(soup)

#중요 정보 싹 뽑기!
#print(soup.find_all(태그명, 속성명))
print(soup.find_all('tbody',id="_topItems4")[0].text) #class 로 찾을때는 예약어가 있기때문에 class_
#띄어쓰기는 두개의 class, 둘 중 하나만 쓰면 됨
```
