```python
score = [90, 25, 67, 45, 80]
number = 0
for mark in score:
  number = number + 1
  if mark >= 60:
    print("%d번 학생은 합격입니다" % number)
  else:
    print("%d번 학생은 불합격입니다" % number)
```

```python
#for문에 continue문 사용
for mark in score:
  number = number + 1
  if mark < 60:
    continue
  print("%d번 학생 축하합니다. 합격입니다." % number)
```

```python
#여러 개의 입력값을 받는 함수
#- def 함수_이름(*매개변수):
    #수행할_문장
def add(*nums):
  result = 0
  for i in nums:
    result = result + i #nums에 입력받은 모든 값을 더한다
  return result

result = add(1,2,3)
print(result)

def add_mul(choice, *args):
  if choice == "add":
    result = 0
    for i in args:
      result = result + i
  elif choice == 'mul':
    result = 1
    for i in args:
      result = result * i
  return result

result = add_mul('add', 1, 2, 3, 4)
print(result)

result = add_mul("mul", 1, 2, 3, 4)
print(result)
```

```python
#변수를 정하지 않고 사용자와 interactive하는 함수
def findmin():
     x = int(input('enter first integer: '))
     y = int(input('enter second integer: '))
     z = int(input('enter the last integer: '))
     minimum = x
     if y < minimum:
      minimum = y
     if z < minimum:
      minimum = z
     return minimum
```

```python
##python standard library

#random 함수 연습
import random
frequency0 = 0
frequency1 = 0
sample = 10_000 #숫자 천단위 구분은 underbar을 이용한다

for roll in range(sample):
    face = random.randrange(0,2)
    if face == 0:
        frequency0 += 1
    else:
        frequency1 += 1

print(f'{"face"}{"freqeuncy":>13}{"ratio":>10}')
print(f'{0:>4}{frequency0:>13}{frequency0/sample*100:>10.2f}')
print(f'{1:>4}{frequency1:>13}{frequency1/sample*100:>10.2f}')


```

```python
#내가 직접 만든 random dice count & string code ^_^
import random as rd

numb1 = 0
numb2 = 0
numb3 = 0
numb4 = 0
numb5 = 0
numb6 = 0
sample = 30
for roll in range(sample):
    dice = rd.randrange(1, 7)
    if dice == 1:
        numb1 += 1
    elif dice == 2:
        numb2 += 1
    elif dice == 3:
        numb3 += 1
    elif dice == 4:
        numb4 += 1
    elif dice == 5:
        numb5 += 1
    else:
        numb6 += 1

print(" <Let's count the dice~ ^_^ >\n")

print(f'Dice number{"Count":>8}     {"count as ★"}\n')
print(f'{1:>5}{numb1:>14}     {"★" * numb1}')
print(f'{2:>5}{numb2:>14}     {"★" * numb2}')
print(f'{3:>5}{numb3:>14}     {"★" * numb3}')
print(f'{4:>5}{numb4:>14}     {"★" * numb4}')
print(f'{5:>5}{numb5:>14}     {"★" * numb5}')
print(f'{6:>5}{numb6:>14}     {"★" * numb6}')
print("-" * 41)
print("\n" f'{"Total :":>6}{sample:>12}')
print("\n")
a = "{0:-^40}".format("끝")
print(a)
```
